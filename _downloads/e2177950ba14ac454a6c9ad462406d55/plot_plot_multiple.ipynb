# Using [plot_multiple]{.title-ref} to plot several datasets on the same figure

In this example, we will display several Raman datasets on the same
figure using the [plot_multiple]{.title-ref} helper. Unlike
[multiplot]{.title-ref}, which creates a grid of panels,
[plot_multiple]{.title-ref} overlays several datasets on one shared
axes. Several options are available to customize that overlay.


Import spectrochempy as usual


```python
import os
from pathlib import Path

import spectrochempy as scp
```

Load the data (here 2D spectrum made from a list of 1D spectra):


```python
TEST_FILE = Path(os.environ.get("TEST_FILE", "ramandata/labspec/serie190214-1.txt"))
B1 = scp.read(TEST_FILE)
```

First we show the basic plot. Here [cmap=None]{.title-ref} uses
categorical rotating colors for a line-based plot, and [lw]{.title-ref}
is the short alias for [linewidth]{.title-ref}. We also enlarge the
figure for readability.


```python
_ = B1.plot(cmap=None, lw=1)
```

We will limit the x range to the region of interest note the float
number to specify that we use coordinates and not indices


```python
B2 = B1[:, 60.0:]
```

As there is obviously a drift in these spectra, we will use detrend to
remove it.


```python
B3 = scp.detrend(B2)
_ = B3.plot(cmap=None)
```

To demonstrate the use of [plot_multiple]{.title-ref} we will take only
a few spectra. For instance the 5 first spectra:


```python
B4 = B3[:5]
```

plot it to see what we have selected


```python
_ = B4.plot(cmap=None)
```

Now use [plot_multiple]{.title-ref} to overlay the selected spectra on
one axes. We use [method=\"pen\"]{.title-ref} for line rendering, set
labels for the combined legend, and apply [shift]{.title-ref} so the
traces remain visually separated.


```python
datasets = list(B4)
_ = scp.plot_multiple(
    datasets,
    method="pen",
    legend="best",
    labels=["A", "B", "C", "D", "E"],
    color=["black", "red", "green", "blue", "violet"],
    lw=[1, 2.5, 1, 1, 1],  # line width (we use here different values)
    ls="-",  # solid line style
    shift=1000,  # vertical shift
)
```
